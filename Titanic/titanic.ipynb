```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
```

```python
data1 = pd.read_csv('E:/ANUBHAV/VS Code/ML/Datasets/Data/titanic/train.csv')
data1
```

```python
data2 = data1.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis = 1)
data2
```

```python
data2['Embarked'].fillna('Q', inplace=True)
Counter(data2['Embarked'])
```

```python
data2['Age'].fillna(data2['Age'].mean(), inplace=True)
data2
```

```python
data2['Embarked'].replace('S', 0, inplace=True)
data2['Embarked'].replace('C', 1, inplace=True)
data2['Embarked'].replace('Q', 2, inplace=True)
data2
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
x = np.array(data2.drop(['Survived'], axis=1))
scale = StandardScaler()
x = scale.fit_transform(x)
y = np.array(data2['Survived'])
x, y
```

```python
from sklearn.model_selection import cross_val_score
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
```

```python
print(f'LogisticRegression: {cross_val_score(LogisticRegression(), x, y, cv=10).mean()}')
print(f'Decision Tree: {cross_val_score(DecisionTreeClassifier(), x, y, cv=10).mean()}')
print(f'Random Forest: {cross_val_score(RandomForestClassifier(), x, y, cv=10).mean()}')
print(f'SVM: {cross_val_score(SVC(), x, y, cv=10).mean()}')
print(f'KNN: {cross_val_score(KNeighborsClassifier(), x, y, cv=10).mean()}')
print(f'Naive Bayes: {cross_val_score(GaussianNB(), x, y, cv=10).mean()}')
```

```python
print(cross_val_score(SVC(kernel='poly'), x, y, cv=10).mean())
print(cross_val_score(SVC(kernel='sigmoid'), x, y, cv=10).mean())
```

```python
print(cross_val_score(KNeighborsClassifier(n_neighbors=15, p=2), x, y, cv=10).mean())
```

```python
print(cross_val_score(RandomForestClassifier(n_estimators=200), x, y, cv=10).mean())
```

```python
model = SVC()
model.fit(x, y)
```

```python
test_data = pd.read_csv('E:/ANUBHAV/VS Code/ML/Datasets/Data/titanic/test.csv')
test_data
```

```python
tdata = test_data.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis = 1)
tdata
```

```python
pred = model.predict(x_test)
pred
```
